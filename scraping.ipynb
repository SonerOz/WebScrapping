```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import csv
```

```python


url = "https://www.glassdoor.com/index.htm"

driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)

SignIn = driver.find_element_by_class_name("locked-home-sign-in").click()

file = open('config.txt')
lines = file.readlines()
username = lines[0]
password = lines[1]

Email = driver.find_element_by_name("username")
Email.send_keys(username)
Password = driver.find_element_by_name("password")
Password.send_keys(password)

Submit = driver.find_element_by_name("submit").click()

time.sleep(5)

Button = driver.find_element_by_xpath("//*[@id='scBar']/div/button")
Button.click()
time.sleep(5)

arama = driver.find_element_by_xpath("//*[@id='sc.keyword']")
time.sleep(3)
arama.send_keys("machine learning")
time.sleep(3)
buton = driver.find_element_by_xpath("//*[@id='scBar']/div/button")
buton.click()

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

print(soup)

```

All jobs 

```python
jobs = soup.findAll("li",{"class":"jl react-job-listing gdGrid"})
```

Collect all these infos to the csv file

```python
csvFile = open("Jobs.csv", "w", newline="")

try:
    writer = csv.writer(csvFile)
    writer.writerow(["Rating","EmployerName","Location","Salary","Hiring","Days","URL"])
    for job in jobs:
        try:
            rating = job.find("span",{"class":"compactStars"}).get_text()
        except Exception as e:
            rating = "N/A"
        try:
            EmployerName = job.find("div",{"class":"jobInfoItem jobEmpolyerName"}).get_text()
        except Exception as e:
            EmployerName = "N/A"
        try:
            Location = job.find("span",{"class":"subtle loc css-nq3w9f pr-xxsm"}).get_text()
        except Exception as e:
            Location = "N/A"
        try:
            Salary = job.find("span",{"class":"gray salary"}).get_text()
        except Exception as e:
            Salary = "N/A"
        try:
            Hiring = job.find("div",{"class":"hotListing css-65p68w css-m9i057 css-1uov7ef"}).get_text()
        except Exception as e:
            Hiring = "N/A"
        try:
            Days = job.find("div",{"class":"d-flex align-items-end pl-std minor css-65p68w"}).get_text()
        except Exception as e:
            Days = "N/A"
        try:
            Url = job.a.attrs["href"]
        except Exception as e:
            Url = "N/A"
        writer.writerow([rating, EmployerName, Location, Salary, Hiring, Days, Url])
finally:
    csvFile.close()
```
